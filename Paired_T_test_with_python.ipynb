#Paired T-Test

A paired t-test is used when we are interested in the difference between two variables for the same subject.

paired t test.jpg

```python
# importing libraries
import numpy as np
import scipy.stats as stats
import pandas as pd
```

```python
# Considering weights as variables
w1 = [25,30,28,35,28,34,26,29,30,26,28,32,31,30,45]
w2 = w1+stats.norm.rvs(scale=5,loc=-1.25,size=15)
print(w1)
print(w2)
```

```python
df_weight = pd.DataFrame({"weight_1":np.array(w1), "weight_2":np.array(w2), "weight_change":np.array(w2)-np.array(w1)}) # creating dataframe with weight and its differences.
```

```python
df_weight
```

```python
ttest,p_val = stats.ttest_rel(a=w1,b=w2)
```

```python
if p_val<0.05: # aplha value is 0.05
  print("We are rejecting Null Hypothesis")
else:
  print("We are accepting Null Hypothesis")
```

There is no difference in the variables.
